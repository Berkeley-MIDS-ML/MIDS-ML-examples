First, let's just plot these.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# I am being lazy and leaving pixel numbers, so I get that data
for i, im in enumerate(sample_images.images):
    # Note that subplot breaks with other scipy conventions
    # Numbering starts at 1, so we need to add 1 to our index
    plt.subplot(1,2,i+1)
    plt.imshow(im)
    
```

```python
pagoda = sample_images.images[0]
pagoda.shape
```

We see that these are 3D arrays. Each "pane" (slice along the 3rd or "2th" dimension) represents a color.

```python
# This is now monochrome, so we want to supply a sensible colormap
pagoda_red = pagoda[:,:,0]
plt.imshow(pagoda_red, cmap='binary')
```

```python
# Note that we've dropped the 2th dimension
pagoda_red.shape
```

```python
# Like with our digit data, these are effectively 8-bit values
pagoda_red.min(), pagoda_red.max()
```

```python
# We can create a logical vector like this
pagoda_red > 128
```

```python
# Let's make a new version of this array that we can change
pagoda_thresh = pagoda_red.copy()
```

```python
# I interactively played with the threshold to find a nice value here
pagoda_thresh[pagoda_red > 70] = 1
pagoda_thresh[pagoda_red <= 70] = 0
```

```python
plt.imshow(pagoda_thresh, cmap='binary')
```

```python
pagoda_thresh[:] = 0
pagoda_thresh[pagoda_red > 70] = 1
pagoda_thresh[pagoda_red > 150] = 2
```

```python
# Here, it is reasonable to use the default cmap, as our classes are
# quite clear / distinct
plt.imshow(pagoda_thresh)
```

```python
# But 'binary' is a better comparison to the full monochrome version above
plt.imshow(pagoda_thresh, cmap='binary')
```

Note - you can get into trouble if you forget to apply indexing. The following overwrites the matrix with a single number! (It becomes an `int`)

```python
pagoda_thresh = 0
pagoda_thresh[pagoda_red > 70] = 1
pagoda_thresh[pagoda_red > 150] = 2
```

```python
type(pagoda_thresh)
```

You can fix the above code by adding back the indexing operation, which assigns 0 *into* the numpy array (`pagoda_thresh[:] = 0`)

# Blur

```python
import numpy as np
from scipy.ndimage import filters
```

A convolution will essentially broadcast a multiplication mask operation over an array. Note that for reasons relating to the physical interpretation of convolution, the array will be symmetrically flipped. This won't matter because we'll only be using symmetric kernels.

```python
# I'm using 25 pixels, as 
kernel = np.ones((5,5))
kernel[1:4,1:4] = 2
kernel[2,2] = 2.5
kernel
```

```python
plt.imshow?
```

```python
# A good kernel should sum to one
kernel /= kernel.sum()
# matshow is a shortcut to a reasonable imshow for matrices
plt.matshow(kernel, cmap='binary')
```

```python
# We can also look at the contour through the middle like this
plt.plot(kernel[2,])
```

```python
# Convolve computes the
pagoda_blur1 = filters.convolve(pagoda_red, kernel)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(pagoda_red, cmap='binary')
plt.subplot(1,2,2)
plt.imshow(pagoda_blur1, cmap='binary')
```

```python
# Note the shapes are still the same - edge cases are handled automatically
# By default, values are "reflected" back, which is reasonable for
# the kind of image processing we're doing
pagoda_red.shape, pagoda_blur1.shape
```

The gaussian_filter funtion is easy to use, but doesn't restrict to local pixels

```python
# Note - I'm using a large sigma here, as my image is very high-detail.
# You should use a sigma that gives you a "reasonable" blur
pagoda_blur2 = filters.gaussian_filter(pagoda_red, 3)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(pagoda_red, cmap='binary')
plt.subplot(1,2,2)
plt.imshow(pagoda_blur2, cmap='binary')
```
